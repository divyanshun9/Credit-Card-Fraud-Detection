```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#taking only 492 samples out of 284315 samples of class 1 so that we can train out model accordingly
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# training the RandomForestClassifier Model with Training Data
clf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
clf.fit(X_train, Y_train)
```

```python
# training the DecisionTreeClassifier Model with Training Data
c = DecisionTreeClassifier(max_depth=5, random_state=42)
c.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction_lr = model.predict(X_train)
training_data_accuracy_lr = accuracy_score(X_train_prediction_lr, Y_train)
X_train_prediction_rf = clf.predict(X_train)
training_data_accuracy_rf = accuracy_score(X_train_prediction_rf, Y_train)
X_train_prediction_dt = c.predict(X_train)
training_data_accuracy_dt = accuracy_score(X_train_prediction_dt, Y_train)
```

```python
print('Accuracy on Training data using logistic regression: ', training_data_accuracy_lr)
print('Accuracy on Training data using random forest classifier: ', training_data_accuracy_rf)
print('Accuracy on Training data using decision tree classifier: ', training_data_accuracy_dt)
```

```python
# accuracy on test data
X_test_prediction_lr = model.predict(X_test)
test_data_accuracy_lr = accuracy_score(X_test_prediction_lr, Y_test)

X_test_prediction_rf = clf.predict(X_test)
test_data_accuracy_rf = accuracy_score(X_test_prediction_rf, Y_test)

X_test_prediction_dt = c.predict(X_test)
test_data_accuracy_dt = accuracy_score(X_test_prediction_dt, Y_test)
```

```python
print('Accuracy score on Test Data using logistic regression : ', test_data_accuracy_lr)
print('Accuracy score on Test Data random forest classifier : ', test_data_accuracy_rf)
print('Accuracy score on Test Data decision tree classifier: ', test_data_accuracy_dt)
```
